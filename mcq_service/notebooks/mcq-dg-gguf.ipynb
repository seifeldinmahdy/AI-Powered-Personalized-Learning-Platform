```python
import os, shutil
from unsloth import FastLanguageModel

total, used, free = shutil.disk_usage("/kaggle/working")
print(f"Disk — Total: {total/1e9:.1f} GB  Used: {used/1e9:.1f} GB  Free: {free/1e9:.1f} GB")
```

```python
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="/kaggle/input/datasets/seifmahdy1/mcq-dg-adapter/results (2)/mcq_dg/adapter",
    max_seq_length=512,
    load_in_4bit=True,
)

model.save_pretrained_gguf(
    "/kaggle/working/mcq_dg_gguf",
    tokenizer,
    quantization_method="q4_k_m",
)

print("\n✅ ALL DONE")
!ls -lh /kaggle/working/mcq_dg_gguf/
```
