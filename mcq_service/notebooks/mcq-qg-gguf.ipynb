## 2. Check Disk Space


```python
import os, shutil
from unsloth import FastLanguageModel

total, used, free = shutil.disk_usage("/kaggle/working")
print(f"Disk — Total: {total/1e9:.1f} GB  Used: {used/1e9:.1f} GB  Free: {free/1e9:.1f} GB")
```

## 3. Configuration

Update `ADAPTER_PATH` to the exact path where the trained QG adapter was saved
in your attached Kaggle dataset (look inside the dataset for the `adapter/` folder).


```python
# ── EDIT THIS to match the path of your attached Kaggle dataset ───────────────
# The kaggle_train_qg notebook saves the adapter to:
#   /kaggle/working/mcq_qg/adapter/
# When you attach that notebook's output as a dataset the path becomes:
#   /kaggle/input/<dataset-slug>/mcq_qg/adapter
# Example:
#   ADAPTER_PATH = "/kaggle/input/mcq-qg-adapter/mcq_qg/adapter"
ADAPTER_PATH = "/kaggle/input/datasets/seifmahdy1/mcq-qg-adapter/mcq_qg/adapter"

GGUF_OUTPUT  = "/kaggle/working/mcq_qg_gguf"
MAX_SEQ_LEN  = 1024   # must match the value used during QG training
QUANT        = "q4_k_m"

print(f"Adapter : {ADAPTER_PATH}")
print(f"Output  : {GGUF_OUTPUT}")
print(f"Quant   : {QUANT}")
print(f"Adapter exists: {os.path.isdir(ADAPTER_PATH)}")
```

## 4. Load QG Adapter & Export to GGUF


```python
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=ADAPTER_PATH,
    max_seq_length=MAX_SEQ_LEN,
    load_in_4bit=True,
)

model.save_pretrained_gguf(
    GGUF_OUTPUT,
    tokenizer,
    quantization_method=QUANT,
)

print("\n✅ ALL DONE")
!ls -lh /kaggle/working/mcq_qg_gguf/
```
